A ferromagnet has aligned and parallel spins.

When an object has disordered spins, the different spin orientations result in zero magnetization.

Monte Carlo simulation allows minimizing the system's energy by accepting spin changes that optimize the energy.

```python
import numpy as np 
import matplotlib.pyplot as plt 
import scienceplots 

plt.style.use(['ggplot'])
# Empty list to store energies : 
energy_spot = []


# Attribuation of random energies to spins, between 0 and 1: 
i = 0
while i < 100:
    energy_spot.append(np.random.rand())
    i += 1
# print(energy_spot)
    

# Making the system a 0-1 energy problem :    
for i, value in enumerate(energy_spot):
    if value < 0.5: 
        energy_spot[i] = 0
    elif value >= 0.5:
        energy_spot[i] = 1


# Reshape the energies into a 10x10 grid
energy_grid_10x10 = np.array(energy_spot).reshape(10, 10)
# print(energy_grid_10x10)


# Associating color pink with 0 energy and black with 1 energy: 
colors = np.where(energy_grid_10x10 == 0, 'pink', 'black')


# From the 10 * 10 grid, the upper left will become lower left on graph: 
# Finding 0 energies : 
x_zeros, y_zeros = np.where(energy_grid_10x10 == 0)

# Finding 1 energies : 
x_ones, y_ones = np.where(energy_grid_10x10 == 1)


# Associating colors and energy on graph :
plt.scatter(x_zeros, y_zeros, color=colors[energy_grid_10x10 == 0])
plt.scatter(x_ones, y_ones, color=colors[energy_grid_10x10 == 1])


plt.title('Energy representation of spins (E=0 :pink) (E=1 :black)')


# Affichez le graphique
plt.show()
```
